```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2.1 Exploration of real networks

#### Exercise 2.1

```python
G = nx.Graph()

file = open('../data/network1.csv', 'r')


for line in file:
    
    #Don't read the first 2 lines
    if (line[0] != '#') : 
        
        #Split the line into 2 nodes
        nodes = line.split(',')
        
        #Get rid of the '\n' char at the end of each line
        G.add_edge(nodes[0], nodes[1].split('\n')[0])
    
    
file.close()

print("Number of nodes :", len(G.nodes()))

print("Number of edges :", len(G.edges()))
```

#### Exercise 2.2:  Node degree distribution

#### 2.2.1
The kind of plot we think would be most useful for this kind of task is a repartition of the number of nodes (y-axis) with respect to their degree (x-axis). We will thus use a histogram.
#### 2.2.2

```python
plt.figure(figsize=(15,15))
x = nx.degree(G)

plt.xlabel("number of connections", fontsize = 20)
plt.ylabel("number of persons, NORMALIZED", fontsize = 20)


val = list(x.values())

print("The mean of connections per node is : ", np.mean(val))

print("\nThe number of connexions of the 15 most popular nodes:")
print(sorted(val, reverse = True)[:15], '\n')


# We plot the histogram so that its integral is 1 => Closer to an exponential distribution
plt.hist(val, normed = True, bins = 200, log = False )
```

We can see that some of the properties of the distribution are:
* No numbers is negative (obviously)
* It decreases quite rapibly, almost at an exponential rate
* There are 2 very connected nodes (1482 & 1083 connections) that seem to be two "cores" of the graph

#### 2.2.3

When we normalize the graph, we see that the distribution is close to an exponential distribution. We saw that the mean of the number of connection is ~2.75. We can thus approximate this distribution by an exponential random variable with parameter (lambda = 1/2.75)

#### Exercise 2.3: Giant component

```python
sorted_components = sorted(nx.connected_components(G), key = len, reverse=True)
print('Number of connected components : ', len(sorted_components))
sizes = list()
for comp in sorted_components:
    sizes.append(len(comp))

print('We can see that there is a giant component as the biggest component has size', sizes[0])
print('and the second largest component only has size', sizes[1])
```

We can then conclude that there is one, and one only, giant component, which englobes ${12576} \over {13033}$ nodes, which is more than 96% of the graph. 



#### Exercise 2.4: Short paths & Small-worlds

```python
# ======== Run this only once ! ========
shortest_paths_lengths = nx.all_pairs_shortest_path_length(G)
```

```python
items = list(shortest_paths_lengths.items())
#items is a list of tuples, each tuple is defined like this
#items[0] is a tuple (starting_node, dict(key=ending_node, val=number of connections needed to reach starting_node))


# To compute the average
number_of_values = 0
sum = 0

for e in items:
    partial_vals = list(e[1].values())  # e_th tuple of the list, second elem of the tuple, values of this dictrionnary
    sum += np.sum(partial_vals)
    number_of_values += len(partial_vals)


print("The mean of all connections between two nodes is ", float(sum/number_of_values))
```

We think it is effectively a small-world, since the mean of the degrees of separation is inferior to 6.
