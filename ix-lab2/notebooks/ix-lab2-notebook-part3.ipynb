##### Usefull functions :

```python
'''
Inside loops, this function prints the state, ie the percentage of iterations already accomplished
'''

def print_state(actual_i, modulo, iterations):

    if (actual_i % modulo == 0) :
        print ("\r", 100*float(actual_i)/iterations, " %", end='')
```

```python
'''
Reads and creates a Graph from a json file =
'''
def read_json_file(filename):
    
    with open(filename) as f:
        d = json.load(f)
        
    return json_graph.node_link_graph(d, directed = d["directed"], multigraph = d["multigraph"])
```

#### 2.9.2 Evolution over time

```python
#======== Initialization of 2 usefull lists ========


inf = sir.inf_time # Contains the infection-time of each node
death = sir.rec_time # Contains the death-time of each node
```

```python
# Store the percentage of Susceptible, Infected and Recovered people

nbS = np.array(())
nbI = np.array(())
nbR = np.array(())
```

```python
# ====== Computes the percentage of Susceptible, Infected and Recovered(dead) subject through time ======

time = np.linspace(0, 101, 1000)

n = float(len(inf))

for t in time :
    
    counterS = 0
    counterI = 0
    counterR = 0
    
    for i in inf : 
        
        #At thime t, the subject is not yet infected => he is susceptible
        if (t < i):
            counterS = counterS +1
            
    for d in death :
        
        #At thime t, the subject is already dead => he is recovered        
        if (t > d):
            counterR = counterR +1
    
    # The subject is not recovered, nor susceptible
    counterI = n - counterR - counterS 
    
    nbS = np.append(nbS, counterS)
    nbI = np.append(nbI, counterI)
    nbR = np.append(nbR, counterR)
    
nbS = 100*nbS / n
nbI = 100*nbI / n
nbR = 100*nbR / n
```

```python
#======== PLots the values computed in the cell above ========

plt.figure(figsize=(10, 10))

plt.plot(time, nbS)
plt.plot(time, nbI)
plt.plot(time, nbR)

plt.title("Evolution of the epidic through time",fontsize=18)
plt.xlabel("time", fontsize=15)
plt.ylabel("Percentage", fontsize=15)
unused = plt.legend(["Susceptible", "Infected", "Recovered (dead)"], loc='lower right', fontsize=14)
```

```python
#======== Draws the epidemics ========

    # Green for healthy people
    # Yellow for infected ones
    # Red for recovered ones

def draw_epidemics (graph, sir, time) :
    
    
    #===== Assigning Color =====
    
    color_map =[]
    for i in range(0, graph.order()):
        
        if sir.get_node_status(i,time) == 0:
            color_map.append('green')
            
        elif sir.get_node_status(i, time) ==1:
            color_map.append('yellow')
            
        else:
            color_map.append('red')
            
    
    #===== Plot Parameters =====

    plt.figure(figsize = (10, 10))
    plt.axis('off')
    
    title = str('Evolution of the epidemics after ' + str(time) + ' day(s)')
    plt.title(title)
    
    #legend
    green_patch = mpatches.Patch(color='green', label='Susceptible (healty) people')    
    yellow_patch = mpatches.Patch(color='yellow', label='Infected people')
    red_patch = mpatches.Patch(color='red', label='Recovered (dead) people')
    
    plt.legend(handles=[green_patch, yellow_patch, red_patch])
    
    
    #===== Final Plot =====
    
    nx.draw_networkx(graph, pos, node_color = color_map, node_size = 15, width = 0, with_labels = False)
```

### 2.3.1 Stop the apocalypse!



```python
''' 
Computes and prints the average percentage of S,I and R people after an epidemic in a graph

args :
    - The basic graph that we start from (NewYork roads for us)
    - strategy, the function of edge suppression (either delete_random or smarter_strategy)
    - the number of nodes to be deleted by this strategy
    - the simulation time
    - the number of simulations for our mean

'''

def get_average_status (basic_graph, strategy, nb_nodes_to_delete=2500, simulation_time=30, nb_simulations=50 ) :
    
    order = G.order()
    
    sumS = 0 #susceptible
    sumI = 0 #infected
    sumR = 0 #recovered
    
    # ======== Simulation of 'nb_simulations' cases ========
    for j in range (nb_simulations) : 
        
        # inner counters
        inner_s = 0
        inner_i = 0
        inner_r = 0

        # ======== Inside this with-statement, functions are not allowed to print ========
        with io.capture_output() as captured:
            
           
    
            # Create a modified graph to avoid the epidemic expansion
            newG = strategy(basic_graph, nb_nodes_to_delete)
            new_sir = epidemics_helper.SimulationSIR(newG, 10.0, 0.1)

            #launch the epidemic
            random_node = np.random.choice(newG.nodes())
            new_sir.launch_epidemic(random_node, simulation_time)   
        
        # ========== no print-ban anymore ==========

        

        # ========== Computation ==========    
        
        for s in new_sir.status:
            if s == 0 :
                inner_s += 1
            elif s == 1 :
                inner_i += 1
            else :
                inner_r += 1
    
    
        # Compute the inner percentage of S,I,R people
        
        sumS += inner_s*100.0 / order
        sumI += inner_i*100.0 / order
        sumR += inner_r*100.0 / order
        

        print_state(j, 1, nb_simulations)


    print("\r100 %")

    print("\n\nAfter ", simulation_time, " days and after deleting", nb_nodes_to_delete ," nodes, there are in average :")
    print( float(sumS) / nb_simulations, " % of healthy people")
    print( float(sumI) / nb_simulations, " % of infected people")
    print( float(sumR) / nb_simulations, " % of recovered(dead) people")    
```

#### Exercise 2.10: Strategy 1

The first 'naive' strategy is defined in the cell below :

```python
'''
This function deletes n edges at random in the graph G
It returns the modified graph
'''

def delete_random(G, n):
    
    newG = G.copy()
    new_nodes = newG.nodes()
    
    
    for k in range(n):
        
        #take a random node
        node_1 = np.random.choice(new_nodes)
        nei = newG.neighbors(node_1)
        
        #if this node still has at least one neighbor
        if (len(nei) > 0):
            
            #remove a random edge between node_1 and one of its neighbors
            node_2 = np.random.choice(nei)
            newG.remove_edge(node_1, node_2)
            
        #otherwise, we decrement i because we did not delete any edge during this iteration    
        else :
            k -= 1
        
        print_state(k, 13, n)

    
    print("\r100 %")
        
    return newG
```

#### Exercise 2.11: Strategy 2

Our strategy below uses the fact that some nodes are only the source of 2 edges.

Those nodes are thus critical points because if we cut one of those 2 edges, the epidemic cannot evolves through this point anymore. 

So we begin by destroying one connection for every node that has 2 edges.
Then if we have some 'iteration' left, we apply the algorithm derived from the friends paradox that we studied in class.

##### Extracting nodes with only 2 neighbors (bottleneck nodes)

```python
# ======== Init ========
two_neighbors_nodes = np.array(())
all_other_nodes = np.array(())

# ======== Filling the first list ========
for nd in G.nodes():
    if (len(G.neighbors(nd)) == 2):
        two_neighbors_nodes = np.append(two_neighbors_nodes, nd)
        
# ======== Filling the second list ========
all_other_nodes = np.setdiff1d(G.nodes(), two_neighbors_nodes, assume_unique=True)
```

```python
'''
This function deletes n edges at random in the graph G
It returns the modified graph
'''

def smarter_strategy(G, n):
    
    #To be returned
    newG = G.copy()
    
    for i in range (n): 
        
        
        # ========== Deleting bottleneck edges ==========
        
        if (i < len(two_neighbors_nodes)) :
            
                        
            # Deleting one edge of a node with 2 edges only (like a bridge !)
            n = two_neighbors_nodes[i]
            nei = newG.neighbors(n)
            
            if (len(nei) != 0) : #Could happen if we already deleted the edges of this point
                newG.remove_edge(n, nei[0])
                
        
        # ========== Applying the friends paradox (level 2) ==========
        
        else:
        
            # Choose a first node
            node_1 = np.random.choice(all_other_nodes)
            neighbors = newG.neighbors(node_1)


            if (len(neighbors) != 0) :
                
                # Choose a second node
                node_2 = np.random.choice(neighbors)    
                neighbors_2 = newG.neighbors(node_2)

                if (len(neighbors_2) != 0) :
                    
                    # Choose a third node
                    node_3 = np.random.choice(neighbors_2) 
                    
                    # Delete the edge (node_2, node_3)
                    newG.remove_edge(node_2, node_3)
        
        
        print_state(i, 13, n)        
    
    print("\r100 %")
    return newG
```

```python
#======== Apply our new strategy =========

G3 = smarter_strategy(G,2500)
sir3 = epidemics_helper.SimulationSIR(G3, 10.0, 0.1)

random_node = np.random.choice(G3.nodes())
sir3.launch_epidemic(random_node, 30)
```
